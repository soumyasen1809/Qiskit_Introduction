```python
from qiskit.algorithms import Shor
from qiskit.utils import QuantumInstance
from qiskit import Aer, QuantumCircuit, execute
from qiskit.tools.visualization import plot_histogram
import numpy as np
```

```python
my_shor = Shor(quantum_instance=QuantumInstance(Aer.get_backend('qasm_simulator'), shots=1024))
```

```python
a, N = 2, 15
circuit = Shor.construct_circuit(my_shor, N=N, a=a)
circuit.draw(output='mpl')
```

```python
factors = Shor.factor(my_shor,N=N, a=a)
print(factors)
```
