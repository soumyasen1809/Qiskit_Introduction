```python
import qiskit
from qiskit.tools.visualization import plot_bloch_multivector
%matplotlib inline
from qiskit.tools.visualization import plot_histogram
```

```python
#Build the circuit
qr = qiskit.QuantumRegister(1)
cr = qiskit.ClassicalRegister(1)
circuit = qiskit.QuantumCircuit(qr, cr)

#Apply a Hadamard gate
circuit.x(qr[0])

#Simulate the circuit
result = qiskit.execute(circuit, backend=qiskit.Aer.get_backend('statevector_simulator')).result()
statevector = result.get_statevector()
print(statevector)
circuit.draw(output='mpl')
```

```python
plot_bloch_multivector(statevector)
```

```python
#Plot the state vector
circuit.measure(qr, cr)
result = qiskit.execute(circuit, backend=qiskit.Aer.get_backend('qasm_simulator'), shots=1024).result()
plot_histogram(result.get_counts(circuit))
```

```python
#Get the matrix representation
qr = qiskit.QuantumRegister(1)
cr = qiskit.ClassicalRegister(1)
circuit = qiskit.QuantumCircuit(qr, cr)

#Apply a Hadamard gate
circuit.x(qr[0])

#Get the unitary matrices of the circuit
result = qiskit.execute(circuit, backend=qiskit.Aer.get_backend('unitary_simulator')).result()
unitarymatrix = result.get_unitary()
print(unitarymatrix)
```

```python
plot_bloch_multivector(unitarymatrix)
```
